```python
# Eliminar los caracteres M, B y K de la columna de Volumen
def limpiar_volumen(Volumen):
    if isinstance(Volumen, str):
        Volumen = Volumen.strip()
        if 'M' in Volumen:
            return float(Volumen.replace('M', ''))
        elif 'B' in Volumen:
            return float(Volumen.replace('B', '')) * 1000  # Convert billions to millions
        elif 'K' in Volumen:
            return float(Volumen.replace('K', '')) / 1000  # Convert thousands to millions
    return float(Volumen)

df_BTC['Volumen_BTC'] = df_BTC['Volumen_BTC'].apply(limpiar_volumen)
```

####**7.1.1.1) DXY**

Para graficar las series temporales se calculo la media de la suma del precio de apertura y de cierre para cada fila. Luego se creo una copia del dataframe, para no modificar el original. Se seleccionaron las columnas de medias y fecha para ser graficadas con un gráfico lineal.

Se cálculo el test de mannkendall que permite saber la tendencia historico de los activos y la pendiente de sen para saber la magnitud de la pendiente de los datos. También el porcentaje de aumento del activo desde la fecha inicial hasta la actualidad.

```python
#Libreria de trabajo para el calculo del test de mannkendall
from scipy.stats import norm, kendalltau
```

```python
# Calcular la media del DXY para la columna de precio de cierre
df_unificado.loc[:, 'Media_DXY'] = (df_unificado['Apertura_dolar'] + df_unificado['Cierre_dolar']) / 2
```

```python
# Crear una copia del dataframe
df_copia = df_unificado.copy()
```

Selección de columnas de trabajo para el cálculo del Tau de kendall

```python
# Seleccionar la columna de datos y las fechas
data = df_copia['Media_DXY']
fechas = df_copia['Fecha']
```

Cálculo del Tau de kendall

El tau de kendall calcula la medida de correlación no parámetrica que evalua la relación de orden entre dos variables que indican la tendencia histórica del valor de la media de los precios entre la apertura y de cierre

```python
# Calcular Tau de Kendall y p-valor usando kendalltau
tau, p_value = kendalltau(range(len(data)), data)

# Calcular Z manualmente
n = len(data)
S = 0
for i in range(n - 1):
    for j in range(i + 1, n):
        diff = data.iloc[j] - data.iloc[i]
        S += np.sign(diff)

# Calcular la varianza de S
unique_vals, counts = np.unique(data, return_counts=True)
ties = counts[counts > 1]
var_s = (n * (n - 1) * (2 * n + 5) - np.sum(ties * (ties - 1) * (2 * ties + 5))) / 18

# Calcular Z
if S > 0:
    Z = (S - 1) / np.sqrt(var_s)
elif S < 0:
    Z = (S + 1) / np.sqrt(var_s)
else:
    Z = 0
```

Cálculo de la pendiente de sen

La pendiente de sen calcula la magnitud del tau de kendall

```python
# Calcular la pendiente de Sen
def sens_slope(y):
    slopes = []
    for i in range(len(y) - 1):
        for j in range(i + 1, len(y)):
            slopes.append((y[j] - y[i]) / (j - i))
    return np.median(slopes)

sen_slope = sens_slope(data.values)

# Crear un DataFrame para mostrar los resultados
resultados = pd.DataFrame({
    'Métrica': [
        'Tau de Kendall',
        'p-valor',
        'Estadístico S',
        'Varianza de S',
        'Valor Z',
        'Pendiente de Sen'
    ],
    'Valor': [
        tau,
        p_value,
        S,
        var_s,
        Z,
        sen_slope
    ]
})

# Configuración de la visualización
resultados.set_index('Métrica', inplace=True)
print(resultados)
```

**Resultados de la estadística**

Existe una correlación positiva moderada a fuerte en la serie temporal del precio medio histórico del DXY entre 2010 y 2025, con un coeficiente Tau de Kendall de 0.635. Según el valor p (2.35e-36), esta tendencia es estadísticamente significativa, lo que indica que el crecimiento observado no es producto del azar. La varianza de S (631866.3) sugiere una alta volatilidad en los cambios del precio del DXY, a pesar de la tendencia positiva. Además, el valor Z (12.59) es considerablemente alto y se aleja de cero, confirmando que la tendencia observada no es aleatoria y valida el comportamiento ascendente. Finalmente, la pendiente de Sen (0.160) mide el incremento en la media del DXY por unidad de tiempo, lo que refleja un crecimiento constante y sostenido a un ritmo moderado en la serie temporal.

```python
# Asegurándonos de que las fechas sean del tipo datetime
fechas = pd.to_datetime(fechas)

# Convertir fechas a números ordinales
fechas_ordinales = fechas.map(lambda x: x.toordinal())

# Ajuste lineal
coefs = np.polyfit(fechas_ordinales, data, 1)  # [pendiente, intersección]
linea_tendencia = np.poly1d(coefs)

# Generar valores de la línea de tendencia
y_tendencia = linea_tendencia(fechas_ordinales)
```

```python
# Calcular el porcentaje de aumento
valor_inicial = y_tendencia[0]
valor_final = y_tendencia[-1]
porcentaje_aumento = ((valor_final - valor_inicial) / valor_inicial) * 100

# Mostrar el porcentaje de aumento
print(f"Porcentaje de aumento DXY: {porcentaje_aumento:.2f}%")
```

El aumento de la serie temporal de la media de DXY entre 2010-2025 es de 35.74%, lo que indica que el DXY incrementó su valor en un 35.74% respecto a su valor inicial en 2010

```python
# Calcular Tau de Kendall y p-valor usando kendalltau
tau, p_value = kendalltau(range(len(data)), data)

# Calcular Z manualmente
n = len(data)
S = 0
for i in range(n - 1):
    for j in range(i + 1, n):
        diff = data.iloc[j] - data.iloc[i]
        S += np.sign(diff)

# Calcular la varianza de S
unique_vals, counts = np.unique(data, return_counts=True)
ties = counts[counts > 1]
var_s = (n * (n - 1) * (2 * n + 5) - np.sum(ties * (ties - 1) * (2 * ties + 5))) / 18

# Calcular Z
if S > 0:
    Z = (S - 1) / np.sqrt(var_s)
elif S < 0:
    Z = (S + 1) / np.sqrt(var_s)
else:
    Z = 0
```

Cálculo de la pendiente de sen

```python
# Calcular la pendiente de Sen
def sens_slope(y):
    slopes = []
    for i in range(len(y) - 1):
        for j in range(i + 1, len(y)):
            slopes.append((y[j] - y[i]) / (j - i))
    return np.median(slopes)

sen_slope = sens_slope(data.values)

# Crear un DataFrame para mostrar los resultados
resultados = pd.DataFrame({
    'Métrica': [
        'Tau de Kendall',
        'p-valor',
        'Estadístico S',
        'Varianza de S',
        'Valor Z',
        'Pendiente de Sen'
    ],
    'Valor': [
        tau,
        p_value,
        S,
        var_s,
        Z,
        sen_slope
    ]
})

# Configuración de la visualización
resultados.set_index('Métrica', inplace=True)
print(resultados)
```

**Resultados de la estadística**

Existe una correlación positiva muy fuerte en la serie temporal del precio medio histórico del BTC entre 2010 y 2025, con un coeficiente Tau de Kendall de 0.851. Este valor sugiere un crecimiento sostenido y ordenado del precio del BTC en ese periodo. Según el valor p (9.05e-64), esta tendencia es estadísticamente significativa, lo que indica que el crecimiento observado no es producto del azar, sino que responde a un patrón claro y estructurado en el mercado. La varianza de S (631868.3) muestra un alto nivel de dispersión en los datos, lo cual refleja una fuerte volatilidad en el precio del BTC. Esta volatilidad es característica del comportamiento del activo, influenciada por eventos de adopción masiva, regulación y crisis económicas globales. El valor Z (16.85) es extremadamente alto y se aleja significativamente de cero, lo que valida que la tendencia observada no es aleatoria y refuerza la hipótesis de un comportamiento ascendente constante. Finalmente, la pendiente de Sen (210.65) es considerablemente alta, reflejando un crecimiento robusto y acelerado en el tiempo.
