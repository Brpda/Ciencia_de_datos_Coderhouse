####**7.1.1.1) DXY**

Para graficar las series temporales se calculo la media de la suma del precio de apertura y de cierre para cada fila. Luego se creo una copia del dataframe, para no modificar el original. Se seleccionaron las columnas de medias y fecha para ser graficadas con un grfico lineal.

Se clculo el test de mannkendall que permite saber la tendencia historico de los activos y la pendiente de sen para saber la magnitud de la pendiente de los datos. Tambin el porcentaje de aumento del activo desde la fecha inicial hasta la actualidad.

Finalmente se grfico las series temporales junto con sus lineas de tendencias.

```python
#Libreria de trabajo para el calculo del test de mannkendall
from scipy.stats import norm, kendalltau
```

Se calcula la media entre el precio de apertura y el del cierre para analizar su evolucin temporal.

```python
# Calcular la media del DXY para la columna de precio de cierre
df_unificado.loc[:, 'Media_DXY'] = (df_unificado['Apertura_dolar'] + df_unificado['Cierre_dolar']) / 2
```

```python
# Seleccionar la columna de datos y las fechas
data = df_copia['Media_DXY']
fechas = df_copia['Fecha']
```

Clculo del Tau de kendall

El tau de kendall calcula la medida de correlacin no parmetrica que evalua la relacin de orden entre dos variables que indican la tendencia histrica del valor de la media de los precios entre la apertura y de cierre

```python
# Calcular Tau de Kendall y p-valor usando kendalltau
tau, p_value = kendalltau(range(len(data)), data)

# Calcular Z manualmente
n = len(data)
S = 0
for i in range(n - 1):
    for j in range(i + 1, n):
        diff = data.iloc[j] - data.iloc[i]
        S += np.sign(diff)

# Calcular la varianza de S
unique_vals, counts = np.unique(data, return_counts=True)
ties = counts[counts > 1]
var_s = (n * (n - 1) * (2 * n + 5) - np.sum(ties * (ties - 1) * (2 * ties + 5))) / 18

# Calcular Z
if S > 0:
    Z = (S - 1) / np.sqrt(var_s)
elif S < 0:
    Z = (S + 1) / np.sqrt(var_s)
else:
    Z = 0
```

Clculo de la pendiente de sen

La pendiente de sen calcula la magnitud del tau de kendall

```python
# Calcular la pendiente de Sen
def sens_slope(y):
    slopes = []
    for i in range(len(y) - 1):
        for j in range(i + 1, len(y)):
            slopes.append((y[j] - y[i]) / (j - i))
    return np.median(slopes)

sen_slope = sens_slope(data.values)

# Crear un DataFrame para mostrar los resultados
resultados = pd.DataFrame({
    'Mtrica': [
        'Tau de Kendall',
        'p-valor',
        'Estadstico S',
        'Varianza de S',
        'Valor Z',
        'Pendiente de Sen'
    ],
    'Valor': [
        tau,
        p_value,
        S,
        var_s,
        Z,
        sen_slope
    ]
})

# Configuracin de la visualizacin
resultados.set_index('Mtrica', inplace=True)
print(resultados)
```

**Resultados de la estadstica**

Existe una correlacin positiva moderada a fuerte en la serie temporal del precio medio histrico del DXY entre 2010 y 2025, con un coeficiente Tau de Kendall de 0.635. Segn el valor p (2.35e-36), esta tendencia es estadsticamente significativa, lo que indica que el crecimiento observado no es producto del azar. La varianza de S (631866.3) sugiere una alta volatilidad en los cambios del precio del DXY, a pesar de la tendencia positiva. Adems, el valor Z (12.59) es considerablemente alto y se aleja de cero, confirmando que la tendencia observada no es aleatoria y valida el comportamiento ascendente. Finalmente, la pendiente de Sen (0.160) mide el incremento en la media del DXY por unidad de tiempo, lo que refleja un crecimiento constante y sostenido a un ritmo moderado en la serie temporal.

```python
# Asegurndonos de que las fechas sean del tipo datetime
fechas = pd.to_datetime(fechas)

# Convertir fechas a nmeros ordinales
fechas_ordinales = fechas.map(lambda x: x.toordinal())

# Ajuste lineal
coefs = np.polyfit(fechas_ordinales, data, 1)  # [pendiente, interseccin]
linea_tendencia = np.poly1d(coefs)

# Generar valores de la lnea de tendencia
y_tendencia = linea_tendencia(fechas_ordinales)
```

```python
# Calcular Tau de Kendall y p-valor usando kendalltau
tau, p_value = kendalltau(range(len(data)), data)

# Calcular Z manualmente
n = len(data)
S = 0
for i in range(n - 1):
    for j in range(i + 1, n):
        diff = data.iloc[j] - data.iloc[i]
        S += np.sign(diff)

# Calcular la varianza de S
unique_vals, counts = np.unique(data, return_counts=True)
ties = counts[counts > 1]
var_s = (n * (n - 1) * (2 * n + 5) - np.sum(ties * (ties - 1) * (2 * ties + 5))) / 18

# Calcular Z
if S > 0:
    Z = (S - 1) / np.sqrt(var_s)
elif S < 0:
    Z = (S + 1) / np.sqrt(var_s)
else:
    Z = 0
```

Clculo de la pendiente de sen

```python
# Calcular la pendiente de Sen
def sens_slope(y):
    slopes = []
    for i in range(len(y) - 1):
        for j in range(i + 1, len(y)):
            slopes.append((y[j] - y[i]) / (j - i))
    return np.median(slopes)

sen_slope = sens_slope(data.values)

# Crear un DataFrame para mostrar los resultados
resultados = pd.DataFrame({
    'Mtrica': [
        'Tau de Kendall',
        'p-valor',
        'Estadstico S',
        'Varianza de S',
        'Valor Z',
        'Pendiente de Sen'
    ],
    'Valor': [
        tau,
        p_value,
        S,
        var_s,
        Z,
        sen_slope
    ]
})

# Configuracin de la visualizacin
resultados.set_index('Mtrica', inplace=True)
print(resultados)
```

**Resultados de la estadstica**

Existe una correlacin positiva muy fuerte en la serie temporal del precio medio histrico del BTC entre 2010 y 2025, con un coeficiente Tau de Kendall de 0.851. Este valor sugiere un crecimiento sostenido y ordenado del precio del BTC en ese periodo. Segn el valor p (9.05e-64), esta tendencia es estadsticamente significativa, lo que indica que el crecimiento observado no es producto del azar, sino que responde a un patrn claro y estructurado en el mercado. La varianza de S (631868.3) muestra un alto nivel de dispersin en los datos, lo cual refleja una fuerte volatilidad en el precio del BTC. Esta volatilidad es caracterstica del comportamiento del activo, influenciada por eventos de adopcin masiva, regulacin y crisis econmicas globales. El valor Z (16.85) es extremadamente alto y se aleja significativamente de cero, lo que valida que la tendencia observada no es aleatoria y refuerza la hiptesis de un comportamiento ascendente constante. Finalmente, la pendiente de Sen (210.65) es considerablemente alta, reflejando un crecimiento robusto y acelerado en el tiempo.
