# **3.Preparación del dataset unificado**

```python
#Renombrar las columnas
df_dolar = df_dolar.rename(columns={'Apertura':'Apertura_dolar','Último':'Cierre_dolar', 'Mínimo':'Mínimo_dolar','Máximo':'Máximo_dolar','Vol.': 'Volumen_dolar', '% var.': 'Variación_%_dolar'})
df_BTC = df_BTC.rename(columns={'Date': 'Fecha', 'Price': 'Cierre_BTC', 'Open': 'Apertura_BTC', 'High': 'Máximo_BTC', 'Low': 'Mínimo_BTC', 'Vol.': 'Volumen_BTC', 'Change %': 'Variación_%_BTC'})
```

```python
#Cambiar de orden las columnas de Cierre y Apertura
nuevo_orden_dolar = ['Fecha','Apertura_dolar','Cierre_dolar','Mínimo_dolar', 'Máximo_dolar', 'Volumen_dolar', 'Variación_%_dolar']
nuevo_orden_BTC = ['Fecha','Apertura_BTC','Cierre_BTC','Mínimo_BTC', 'Máximo_BTC', 'Volumen_BTC', 'Variación_%_BTC']
```

```python
#Nuevo orden para todos los dataset
df_dolar = df_dolar[nuevo_orden_dolar]
df_BTC = df_BTC[nuevo_orden_BTC]
```

```python
# Convertir la columna 'Fecha' de df_dolar al formato año-mes-día
df_dolar['Fecha'] = pd.to_datetime(df_dolar['Fecha'])
df_dolar['Fecha'] = df_dolar['Fecha'].dt.strftime('%Y-%m-%d')

# Convertir la columna 'Fecha' de df_BTC al formato año-mes-día
df_BTC['Fecha'] = pd.to_datetime(df_BTC['Fecha'], dayfirst=True)
df_BTC['Fecha'] = df_BTC['Fecha'].dt.strftime('%Y-%m-%d')
```

```python
#Ordenar los datos desde la fecha más ántigua
df_dolar = df_dolar.sort_values(by ='Fecha', ascending=True)
df_BTC = df_BTC.sort_values(by ='Fecha', ascending=True)
```

```python
# Eliminar filas de fecha entre el enero del 2010 hasta julio del 2010 por índice
df_dolar = df_dolar.drop(index=range(178, 185))
```

En primer lugar para esta etapa de renombrar las columnas para que queden con los mismos nombres. También se transforma el valor de fecha en formato datetime. Posterior se le da un nuevo orden a las columnas para despues aplicarlo a los datasets. Finalmente se ordenan comenzando por las fechas mas antiguas hasta las mas modernas.

**df_dolar**

```python
# Eliminar la columna Volumen en df_dolar
df_dolar.drop('Volumen_dolar', axis=1, inplace=True)
```

```python
#Transformar puntos en comas de las columnas
for col in ['Apertura_dolar', 'Cierre_dolar', 'Máximo_dolar', 'Mínimo_dolar', 'Variación_%_dolar']:
    df_dolar[col] = df_dolar[col].astype(str).str.replace(',', '.', regex=False)

# Eliminar el simbolo de % de la columna de Variación en df_dolar
df_dolar['Variación_%_dolar'] = df_dolar['Variación_%_dolar'].astype(str).str.replace('%', '', regex=False)
df_dolar['Variación_%_dolar'] = pd.to_numeric(df_dolar['Variación_%_dolar'])

# Convertir las columnas a tipo numérico después de eliminar las comas
for col in ['Apertura_dolar', 'Cierre_dolar', 'Máximo_dolar', 'Mínimo_dolar','Variación_%_dolar']:
    df_dolar[col] = pd.to_numeric(df_dolar[col])
```

Especificamente para df_dolar se elimino la columna de Volumen ya que casi todos los datos correspondian a datos tipo NaN. Se transformaron las comas en puntos como separador decimal, ya que es como trabaja python con datos númericos. Tambien se elimino el simbolo de % de la columna de variación.

**df_BTC**

```python
# Eliminar el simbolo de % de la columna de Variación en df_BTC
df_BTC['Variación_%_BTC'] = df_BTC['Variación_%_BTC'].astype(str).str.replace('%', '', regex=False)
df_BTC['Variación_%_BTC'] = pd.to_numeric(df_BTC['Variación_%_BTC'])
```

```python
# Eliminar los caracteres M, B y K de la columna de Volumen
def limpiar_volumen(Volumen):
    if isinstance(Volumen, str):
        Volumen = Volumen.strip()
        if 'M' in Volumen:
            return float(Volumen.replace('M', ''))
        elif 'B' in Volumen:
            return float(Volumen.replace('B', '')) * 1000  # Convert billions to millions
        elif 'K' in Volumen:
            return float(Volumen.replace('K', '')) / 1000  # Convert thousands to millions
    return float(Volumen)

df_BTC['Volumen_BTC'] = df_BTC['Volumen_BTC'].apply(limpiar_volumen)
```

```python
# Eliminar comas de las columnas numéricas en df_BTC
for col in ['Apertura_BTC', 'Cierre_BTC', 'Máximo_BTC', 'Mínimo_BTC']:
    df_BTC[col] = df_BTC[col].astype(str).str.replace(',', '', regex=False)

# Convertir las columnas a tipo numérico después de eliminar las comas
for col in ['Cierre_BTC', 'Apertura_BTC', 'Máximo_BTC', 'Mínimo_BTC']:
    df_BTC[col] = pd.to_numeric(df_BTC[col])
```

Para df_BTC se elimino el simbolo % de la columna de variación y luego se transformo el dato en númerico. Se eliminaron los caracteres M, K y B de la columna de Volumen. Se elimino la , como separador de miles dejando solo el punto como separador decimal para luego transformar todas las columnas en datos númericos.

**Creación de Datafame unificado**

```python
# Unir los dataframes en función de la columna 'Fecha'
df_unificado = pd.merge(df_dolar, df_BTC, on='Fecha', how='inner')
df_unificado
#Descargar df
# Save the DataFrame to an Excel file
df_unificado.to_excel('df_unificado.xlsx', index=False)

# Download the file
from google.colab import files
files.download('df_unificado.xlsx')
```
